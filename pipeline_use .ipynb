```python
import pandas as pd 
import numpy as np 
from sklearn.preprocessing import MinMaxScaler,OneHotEncoder,StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.tree import DecisionTreeClassifier
from sklearn.compose import ColumnTransformer 
```

```python
df=pd.read_csv(r'C:\Users\SDS\Downloads\train.csv')
```

```python
# remove unusal clumn from data set 

df.drop(columns=['PassengerId','Cabin','Fare','Ticket','Name'],inplace=True)
```

```python
df
```

```python
#target column
x=df.drop(columns=['Survived'])
y=df['Survived']
```

```python
# train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=4)
```

```python
y_train.shape
```

```python
#fill the null value 
tf1=ColumnTransformer([('age_imputer ',SimpleImputer(),[2]),
                    ('emabrked_imputer ',SimpleImputer(strategy='most_frequent'),[5])],remainder='passthrough')
```

```python
tf1
```

```python
# one hot encoding
tf2=ColumnTransformer([('sex',OneHotEncoder(sparse=False,handle_unknown='ignore'),[1,5])],remainder='passthrough')
```

```python
tf2
```

```python
#scaling
tf3=ColumnTransformer([('scaling',MinMaxScaler(),slice(0,10))])
```

```python
tf3
```

```python
# algoritham
tf4=DecisionTreeClassifier()
```

```python
tf4
```

```python
pipe=Pipeline([('tf1',tf1),
             ('tf2',tf2),
               ('tf3',tf3),
               ('tf4',tf4)
               
             
              ])
```

```python
pipe.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
y_pred=pipe.predict(x_test)
```

```python
accuracy_score(y_pred,y_test)
```
